## Looped Code for ContrlRecords

```python
# -*- coding: utf-8 -*-
"""
Created on Wed Apr  3 13:28:49 2024

@author: mkim3

"""
import os
import fitz  # If you are using Python 3.8, install pymupdf module

def convert_pdf_to_images(pdf_path, output_folder, dpi=300, crop_percentage=None):
    zoom = dpi / 72
    doc = fitz.open(pdf_path)
    for page_number in range(len(doc)):
        page = doc.load_page(page_number)

        if crop_percentage:
            rect = page.rect  
            x0 = rect.width * crop_percentage[0]
            y0 = rect.height * crop_percentage[1] 
            x1 = rect.width * crop_percentage[2]
            y1 = rect.height * crop_percentage[3]
            page.set_cropbox(fitz.Rect(x0, y0, x1, y1))

        mat = fitz.Matrix(zoom, zoom)  
        pix = page.get_pixmap(matrix=mat) 
        
        output_filename = f"{os.path.splitext(os.path.basename(pdf_path))[0]}_{page_number + 1:03d}.png"
        output_path = os.path.join(output_folder, output_filename)
        pix.save(output_path)
        #print(f"Saved: {output_path}")

def process_pdf_files(pdf_folder):
    for root, dirs, files in os.walk(pdf_folder):
        for file in files:
            if file.endswith(".pdf"):
                pdf_path = os.path.join(root, file)
                output_folder = os.path.splitext(pdf_path)[0]
                if not os.path.exists(output_folder):
                    os.makedirs(output_folder)
                crop_percentage = ((0.6/5.8), (1.69/7.5), (5.3/5.8), (4.28/7.5))  # Crop percentage (left, top, right, bottom)
                convert_pdf_to_images(pdf_path, output_folder, dpi=300, crop_percentage=crop_percentage)

pdf_folder = "C:/Users/Jaber/OneDrive - University of Florida/Educational/Research/FHRT/PROJECT/Data_Results/Shands_Samples/ShandsSamples_PDFs/ControlRecords"
process_pdf_files(pdf_folder)
```
